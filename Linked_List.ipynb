```python
class Node:
  def __init__(self,value):
    self.data = value
    self.next = None
```

```python
a = Node(1)
```

```python
print(a)
```

```python
print(a.data)
```

```python
print(a.next)
```

```python
b = Node(2)
c = Node(3)
```

```python
print(b.data)
print(c.data)
```
